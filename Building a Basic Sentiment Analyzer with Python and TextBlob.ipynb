### Task1:

Write a Python function called `sentimentAnalyzer(text)`. This function takes a text (i.e review) and returns the sentiment as follow:

- Score < -0.2: returns **‘Negative’**
- Score between -0.2 and 0.2: returns **‘Neutral’**
- Score > 0.2: returns **‘Positive’**

The score is the polarity of text as determined by TextBlob. An example of creating a textblob object is:

`blob = TextBlob(text)`

*(Refer to TextBlob’s documentation to learn about determining polarity)*


```python
# Install required packages
%pip install textblob
# Download necessary NLTK corpora for TextBlob
import nltk
nltk.download('brown')
nltk.download('punkt')
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
import warnings
warnings.filterwarnings('ignore')
```

```python
def sentimentAnalyzer(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    
    if polarity < -0.2:
        return 'Negative'
    elif polarity > 0.2:
        return 'Positive'
    else:
        return 'Neutral'

print("Sentiment Analyzer function created successfully!")
```

### Task 2:
Verify that the function does classify the sentiment correctly by passing the following words to the function:


|   Word        | Polarity | Resulting Sentiment |
|:-------------:|:--------:|:-------------------:|
| happy         |   0.8    |      Positive       |
| exciting      |   0.3    |      Positive       |
| good          |   0.7    |      Positive       |
| rich          |  0.375   |      Positive       |
| smile         |   0.3    |      Positive       |
| sad           |  -0.5    |      Negative       |
| disappointed  | -0.75    |      Negative       |
| bad           | -0.699   |      Negative       |
| poor          |  -0.4    |      Negative       |
| anger         |  -0.7    |      Negative       |
| food          |   0      |      Neutral        |
| animal        |   0      |      Neutral        |

```python
test_words = {
    'happy': (0.8, 'Positive'),
    'exciting': (0.3, 'Positive'),
    'good': (0.7, 'Positive'),
    'rich': (0.375, 'Positive'),
    'smile': (0.3, 'Positive'),
    'sad': (-0.5, 'Negative'),
    'disappointed': (-0.75, 'Negative'),
    'bad': (-0.699, 'Negative'),
    'poor': (-0.4, 'Negative'),
    'anger': (-0.7, 'Negative'),
    'food': (0, 'Neutral'),
    'animal': (0, 'Neutral')
}

results = []
for word, (expected_polarity, expected_sentiment) in test_words.items():
    blob = TextBlob(word)
    actual_polarity = blob.sentiment.polarity
    actual_sentiment = sentimentAnalyzer(word)
    results.append({
        'Word': word,
        'Expected Polarity': expected_polarity,
        'Actual Polarity': round(actual_polarity, 3),
        'Expected Sentiment': expected_sentiment,
        'Actual Sentiment': actual_sentiment,
        'Match': '✓' if actual_sentiment == expected_sentiment else '✗'
    })

results_df = pd.DataFrame(results)
print("Task 2: Sentiment Verification Results\n")
print(results_df.to_string(index=False))
print(f"\nSuccess Rate: {sum(results_df['Match'] == '✓')}/{len(results_df)} words classified correctly")
```

### Task 3:
Import the provided dataset into a Pandas DataFrame. Filter the data to only include one product of your choice. The product you select must have at least 1000 reviews. Then, describe the data related to your product. This includes:

- Product name

```python
import kagglehub
import os

path = kagglehub.dataset_download("PromptCloudHQ/amazon-reviews-unlocked-mobile-phones")
print("Path to dataset files:", path)

csv_file = os.path.join(path, "Amazon_Unlocked_Mobile.csv")
df = pd.read_csv(csv_file)

print(f"\nDataset loaded successfully!")
print(f"Total reviews: {len(df)}")
print(f"Columns: {len(df.columns)}")
print("\nFirst few rows:")
print(df.head())
```

```python
product_counts = df['Product Name'].value_counts()
products_with_1000_plus = product_counts[product_counts >= 1000]

print("Products with at least 1000 reviews:\n")
for i, (product, count) in enumerate(products_with_1000_plus.head(10).items(), 1):
    print(f"{i}. {product}: {count} reviews")
```

```python
selected_product = products_with_1000_plus.index[0]
product_df = df[df['Product Name'] == selected_product].copy()

product_df = product_df.dropna(subset=['Reviews'])
product_df['Reviews'] = product_df['Reviews'].astype(str)
product_df['Review_Length'] = product_df['Reviews'].apply(len)

print("\nTASK 3: Product Data Description\n")
print(f"Product Name: {selected_product}")
print(f"Number of rows: {len(product_df)}")
print(f"Number of columns: {len(product_df.columns)}")
print(f"\nReview Length Statistics:")
print(f"  Shortest: {product_df['Review_Length'].min()} characters")
print(f"  Longest: {product_df['Review_Length'].max()} characters")
print(f"  Average: {product_df['Review_Length'].mean():.2f} characters")
```

### Task 4:
Apply the function sentimentAnalyzer(text)to the text column in your dataframe. This should create a new column in the dataframe called (Sentiment) which includes the sentiment for each review. 

```python
print("Applying sentiment analysis...")

product_df['Sentiment'] = product_df['Reviews'].apply(sentimentAnalyzer)
product_df['Polarity'] = product_df['Reviews'].apply(lambda x: TextBlob(x).sentiment.polarity)

print("\nTASK 4: Sentiment Analysis Complete\n")
print("Sentiment distribution:")
print(product_df['Sentiment'].value_counts())
print("\nSample results:")
print(product_df[['Reviews', 'Rating', 'Polarity', 'Sentiment']].head(10))
```

```python
import matplotlib.pyplot as plt
import numpy as np

sentiment_counts = product_df['Sentiment'].value_counts()
colors = ['#2ecc71', '#e74c3c', '#95a5a6']

# Plot 1: Sentiment Distribution Bar Chart
plt.figure(figsize=(10, 6), facecolor='white', dpi=600)
plt.style.use('default')
ax = sentiment_counts.plot(kind='bar', color=colors, edgecolor='black')
plt.title('Sentiment Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Sentiment', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
for i, v in enumerate(sentiment_counts.values):
    plt.text(i, v + 20, str(v), ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.savefig('plot1_sentiment_distribution_bar.png', dpi=600, facecolor='white', bbox_inches='tight')
plt.show()

# Plot 2: Sentiment Distribution Pie Chart
plt.figure(figsize=(8, 8), facecolor='white', dpi=600)
plt.pie(sentiment_counts.values, labels=sentiment_counts.index, autopct='%1.1f%%',
        colors=colors, startangle=90, explode=(0.05, 0.05, 0.05))
plt.title('Sentiment Distribution (%)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('plot2_sentiment_distribution_pie.png', dpi=600, facecolor='white', bbox_inches='tight')
plt.show()

# Plot 3: Sentiment Distribution by Rating
plt.figure(figsize=(10, 6), facecolor='white', dpi=600)
sentiment_by_rating = product_df.groupby(['Rating', 'Sentiment']).size().unstack(fill_value=0)
sentiment_by_rating.plot(kind='bar', stacked=False, color=colors)
plt.title('Sentiment Distribution by Rating', fontsize=14, fontweight='bold')
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Sentiment')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('plot3_sentiment_by_rating.png', dpi=600, facecolor='white', bbox_inches='tight')
plt.show()

# Plot 4: Polarity Distribution Histogram
plt.figure(figsize=(10, 6), facecolor='white', dpi=600)
plt.hist(product_df['Polarity'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(-0.2, color='red', linestyle='--', linewidth=2, label='Negative threshold')
plt.axvline(0.2, color='green', linestyle='--', linewidth=2, label='Positive threshold')
plt.title('Polarity Score Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Polarity Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('plot4_polarity_distribution.png', dpi=600, facecolor='white', bbox_inches='tight')
plt.show()

print("All plots saved successfully with white background at 600 DPI!")
```
